```python
# quality filtering (VSEARCH)... trunclen 400.
%%bash
VSEARCH="vsearch"
for s in *.fastq.gz; do
    echo "Quality filtering for sample: $s"
    "$VSEARCH" --fastq_filter "$s" \
        --fastq_maxee 1 \
        --fastq_trunclen 400 \
        --fastq_maxns 0 \
        --fastaout "${s%.fastq}.filtered.fasta" \
        --fasta_width 0
done
```

Sample dereplication

```python
%%bash
VSEARCH="vsearch"

for s in *.filtered.fasta; do
    echo "Dereplicate at sample level and relabel with sample.n for: $s"
    "$VSEARCH" --derep_fulllength "$s" \
        --strand plus \
        --output "${s%.fasta}.derep.fasta" \
        --sizeout \
        --relabel "${s%.filtered.fasta}." \
        --fasta_width 0
done
```

```python
import seaborn as sns, matplotlib.pyplot as plt

# shannon frame
clean_shannon = res_alpha.copy()
clean_shannon['Location'] = clean_shannon['Location'].str.strip()
clean_shannon['Group']    = clean_shannon['Group'].str.strip()

# inv simpson frame
inv_clean = res_alpha_inv.copy()
inv_clean['Location'] = inv_clean['Location'].str.strip()
inv_clean['Group']    = inv_clean['Group'].str.strip()

grp_order = ['WT_10', 'WT_22', 'HET_10', 'HET_22']
panel_lbls = ['A', 'B', 'C', 'D', 'E', 'F' ]


base = sns.color_palette('pastel', 8)
my_palette = dict(zip(grp_order, [base[i] for i in [2, 0, 3, 4]]))

# figure layout
fig, axes = plt.subplots(
    3, 3, figsize=(15, 16),
    gridspec_kw={'width_ratios':[1,1,1],
                 'height_ratios':[1,1,1]},
    sharey=False)

ax_a, ax_b, ax_c = axes[0]
ax_d, ax_e, ax_f = axes[1]

locs        = ['Faecal', 'Ileum', 'Duodenum']
panel_lbls  = ['A', '', '', 'B', '', '', 'C', '', '']

# shannon boxplots
for ax, loc, label in zip([ax_a, ax_b, ax_c], locs, panel_lbls[:3]):
    sub = clean_shannon[clean_shannon['Location'] == loc]
    sns.boxplot(x='Group', y='shannon',
                hue='Group', hue_order=grp_order,
                order=grp_order, palette=my_palette,
                legend=False, data=sub, ax=ax)

    ax.text(-0.12, 1.08, label, transform=ax.transAxes,
            fontweight='bold', fontsize=14, ha='right', va='top')

    ax.set_title(f'{loc}')
    ax.set_xlabel('')
    if ax is not axes[0,0]: ax.set_ylabel(''); ax.set_ylabel('')

# inv simpson boxplots
for ax, loc, label in zip([ax_d, ax_e, ax_f], locs, panel_lbls[3:]):
    sub = inv_clean[inv_clean['Location'] == loc]
    sns.boxplot(x='Group', y='inv_simpson',
                hue='Group', hue_order=grp_order,
                order=grp_order, palette=my_palette,
                legend=False, data=sub, ax=ax)

    ax.text(-0.12, 1.08, label, transform=ax.transAxes,
            fontsize=14, fontweight='bold', ha='right', va='top')

    ax.set_xlabel('')
    if ax is not axes[1,0]: ax.set_ylabel(''); ax.set_ylabel('')


#-----------------
# PCoA
# pcs_bc  = [pc_bcF,  pc_bcI,  pc_bcD]
# meta_bc = [metadata_bcF, metadata_bcI, metadata_bcD]
pcs_w   = [pc_wF,   pc_wI,   pc_wD]
meta_w  = [metadata_wF, metadata_wI, metadata_wD]

import numpy as np
from matplotlib.patches import Ellipse
from scipy.stats import chi2


def add_confidence_ellipse(x, y, ax, n_std=2, trim=0.05, **kwargs):
    if trim:
        r = np.hypot(x - x.mean(), y - y.mean())
        keep = r <= np.quantile(r, 1 - trim)
        x, y = x[keep], y[keep]
    if x.size < 2: # need at least 2 points or it will be pointless.
        return
    cov = np.cov(x, y)
    if np.linalg.det(cov) == 0:
        return
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    vals, vecs = vals[order], vecs[:, order]

    q = np.sqrt(chi2.ppf(0.95, 2)) # 95%, times 2
    width, height = 2 * q * np.sqrt(vals)
    angle = np.degrees(np.arctan2(*vecs[:,0][::-1]))
    ell = Ellipse(xy=(x.mean(), y.mean()),
                  width=width, height=height,
                  angle=angle, **kwargs)
    ax.add_patch(ell)

for ax, pc, meta, label, loc in zip(axes[2], pcs_w, meta_w,
                                    panel_lbls[6:9], locs):
    sns.scatterplot(x=pc.samples.PC1, y=pc.samples.PC2,
                    hue=meta.Group, style=meta.Location,
                    palette=my_palette, legend=False, ax=ax)

    for g in meta.Group.unique():
        m = meta.Group == g
        add_confidence_ellipse(pc.samples.PC1[m].values,
                               pc.samples.PC2[m].values,
                               ax=ax, edgecolor=my_palette[g],
                               facecolor=my_palette[g], linewidth=1.3, alpha=0.15)

    ax.text(-0.12, 1.08, label, transform=ax.transAxes,
            fontweight='bold', fontsize=14, ha='right', va='top')
    ax.set_xlabel('PCoA1 ({:.1f}%)'.format(pc.proportion_explained[0]*100))
    ax.set_ylabel('PCoA2 ({:.1f}%)'.format(pc.proportion_explained[1]*100))

    # p value
    p_result = permanova_results[loc]
    p_val = p_result['p-value']

    if p_val < 0.001:
      p_text = "p < 0.001"
    else:
      p_text = f"p = {p_val:.3f}"
    ax.text(0.97, 0.1, f"PERMANOVA\n{p_text}", transform=ax.transAxes, fontsize=9,
        verticalalignment='top', horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.7))
# do not touch the labels

plt.tight_layout()
plt.show()
plt.savefig('fig6_diversity.png')
```

### heatmap (phylum)

```python
# heatmap of log2 fold change (ZT22 vs ZT10) in phylum... consider separating?
# normalised for each location and change type (WT vs HET). Z-score

from scipy.stats import zscore
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.cm as cm


# fetch copies instead. avoid crashing
locations_data = {
    'Faecal': merged_df_F.copy() if 'merged_df_F' in locals() and merged_df_F is not None else None,
    'Ileum': merged_df_I.copy() if 'merged_df_I' in locals() and merged_df_I is not None else None,
    'Duodenum': merged_df_ST.copy() if 'merged_df_ST' in locals() and merged_df_ST is not None else None
}

# load and calculate
processed_data = {}
all_phyla = set()

# group names
wt_groups = ['WT_10', 'WT_22']
het_groups = ['HET_10', 'HET_22']
required_groups = wt_groups + het_groups

for loc_name, loc_df in locations_data.items():
    if loc_df is None or loc_df.empty:
        print(f"Skipping {loc_name}: No data.")
        continue

    loc_df['Value'] = pd.to_numeric(loc_df['Value'], errors='coerce') # numeric value

    # mean abundance per phylum/group
    mean_abund = loc_df.groupby(['phylum', 'Group'])['Value'].mean().unstack(fill_value=0)

    # check if all groups are loaded... to prevent crashing...
    if not all(g in mean_abund.columns for g in required_groups):
         print(f"Skipping {loc_name}: Missing required groups {required_groups}. Found: {mean_abund.columns.tolist()}")
         continue

    # log2 transform, pseudocount 1
    log2_abund = mean_abund.apply(lambda x: np.log2(x + 1))

    # log2 fold change ZT22 - ZT10
    delta_df = pd.DataFrame({
        'WT_change': log2_abund['WT_22'] - log2_abund['WT_10'],
        'HET_change': log2_abund['HET_22'] - log2_abund['HET_10']
    }, index=log2_abund.index)

    processed_data[loc_name] = delta_df
    all_phyla.update(delta_df.index)

# aligning phyla across all locations + calculate Z-scores
aligned_zscored_data = {}
global_min_max = {'min': float('inf'), 'max': float('-inf')}

sorted_phyla = sorted(list(all_phyla))
for loc_name, delta_df in processed_data.items():
    # re-index to ensure all phyla are present, fill missing.
    aligned_df = delta_df.reindex(sorted_phyla, fill_value=np.nan)
    # calculate Z-score per column - WT_change, HET_change across phyla
    zscored_df = aligned_df.astype(float).apply(zscore, axis=0, nan_policy='omit')
    aligned_zscored_data[loc_name] = zscored_df

    # update global min/max for the colouration
    current_min, current_max = np.nanmin(zscored_df.values), np.nanmax(zscored_df.values)
    if not np.isnan(current_min): global_min_max['min'] = min(global_min_max['min'], current_min)
    if not np.isnan(current_max): global_min_max['max'] = max(global_min_max['max'], current_max)

# handle no min/max + equal min/max
vmin = global_min_max['min'] if np.isfinite(global_min_max['min']) else -1
vmax = global_min_max['max'] if np.isfinite(global_min_max['max']) else 1
if vmin == vmax: vmin -= 0.5; vmax += 0.5

# plotting here
plot_order = ['Faecal', 'Ileum', 'Duodenum']
valid_plots = [loc for loc in plot_order if loc in aligned_zscored_data]
num_plots = len(valid_plots)

if num_plots == 0:
    print("error: no data to plot.")
else:
    # fitting figure size
    num_phyla = len(sorted_phyla)
    num_conditions = 2
    cell_width_inches = 0.3
    fig_width = max(8, cell_width_inches * num_phyla + 2)
    fig_height = cell_width_inches * num_conditions * num_plots + 2.5

    fig = plt.figure(figsize=(fig_width, fig_height))
    gs = gridspec.GridSpec(num_plots, 1, hspace=0.05, top=0.90, bottom=0.25) # vertical spacing

    cmap = sns.diverging_palette(240, 10, as_cmap=True)
    cmap = cmap.copy()
    cmap.set_bad(color='lightgrey') # for NaN values earlier

    ax1 = None # share x-axis

    for i, loc_name in enumerate(valid_plots):
        ax = fig.add_subplot(gs[i, 0], sharex=ax1)
        if i == 0: ax1 = ax # set the axis

        data_to_plot = aligned_zscored_data[loc_name].T # transposing for phyla
        sns.heatmap(data_to_plot, cmap=cmap, center=0, vmin=vmin, vmax=vmax,
                    annot=False, linewidths=.5, ax=ax, cbar=False)

        ax.set_ylabel(loc_name, rotation=0, ha='right', va='center', labelpad=15, fontweight='bold')
        ax.set_title('')
        plt.setp(ax.get_yticklabels(), rotation=0) # y as horizontal

        if i == num_plots - 1:
            plt.setp(ax.get_xticklabels(), rotation=90, visible=True)
            ax.set_xlabel('')
        else:
            plt.setp(ax.get_xticklabels(), visible=False)
            ax.set_xlabel('')

        ax.set_aspect('equal', adjustable='box')

    # add colour bar
    cbar_ax = fig.add_axes([0.15, gs.top + 0.02, 0.7, 0.02]) # no overlap with the plot...
    sm = cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm.set_array([])
    cbar = fig.colorbar(sm, cax=cbar_ax, orientation='horizontal')
    cbar.ax.xaxis.set_ticks_position('top')
    cbar.ax.xaxis.set_label_position('top')

    plt.show()

#---------------- if crash separate.. or remove

import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from PIL import Image

fig = plt.gcf()

fig.set_size_inches(11, 4)

out_png = "fig7_phylum_change_heatmap.png"

fig.canvas.draw()

fig.savefig(out_png,dpi=300,bbox_inches="tight",facecolor="white",format="png")

plt.close(fig)

# Image.open(out_png).verify()

```
